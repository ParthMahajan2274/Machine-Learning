```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
xpoints=np.array([1,2,3,4])
ypoints=np.array([2,4,6,8])
plt.plot(xpoints,ypoints,'ro--',label="2x")
plt.title('My Graph')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.show()
```
