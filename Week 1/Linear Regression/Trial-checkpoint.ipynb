```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_excel('Training data.xlsx')
x_train = np.array(data.iloc[:,0:8])
y_train = np.array(data.iloc[:,8]).reshape(-1,1)
plt.scatter(x_train[:,7],y_train)
# plt.scatter(x_train[:,6],y_train)
# plt.scatter(x_train[:,5],y_train)
# plt.scatter(x_train[:,4],y_train)
# plt.scatter(x_train[:,3],y_train)
# plt.scatter(x_train[:,2],y_train)
# plt.scatter(x_train[:,1],y_train)
# plt.scatter(x_train[:,0],y_train)


# linear iq

# internet	sex	traveltime	studytime	freetime	absences	age	iq

def feature_changing(x_train):
    for i in range(len(x_train[:,0])):
        x_train[i,0] = 0 if (x_train[i,0]=="no") else 1
    for i in range(len(x_train[:,1])):
        x_train[i,1] = 0 if (x_train[i,1]=="M") else 1
    return x_train

x_train = feature_changing(x_train)
x_mean =[]
for i in range(8):
        x_mean.insert(i,float(x_train[:,i].mean()))
        print(x_mean[i])
print(x_mean)
    
```

```python
sum2 = []
variance = []
x_std = []
for i in range(8):
    for j in range((len(x_train[:,i]))):
        sum1=0
        sum1 += (x_train[j,i]-x_mean[i])**2
    sum2.insert(i,sum1)
    variance.insert(i,(sum2[i] / (len(x_train[:,i]))))
    x_std.insert(i,variance[i]**0.5)
def z_score(x_train):
    # x_mean = []
    # x_std=[]
    # sum2 = []
    # variance = []
    # for i in range(8):
    #     x_mean.insert(i,float(x_train[:,i].mean()))
    #     print(x_mean[i])
    # for i in range(8):
    #     for j in range((len(x_train[:,i]))):
    #         sum1=0
    #         sum1 += (x_train[j,i]-x_mean[i])**2
    #     sum2.insert(i,sum1)
    #     variance.insert(i,(sum2[i] / (len(x_train[:,i]))))
    #     x_std.insert(i,variance[i]**0.5)
    for i in range(8):
            x_train[:,i] = ( x_train[:,i] - x_mean[i] ) / x_std[i]
    return x_train,x_std,x_mean
```

```python
def cost(x_train,y_train,w,b):
    loss = 0
    for i in range (1000):
        loss += (y_train[i] - (np.dot(x_train[i],w) + b))**2 / 2000
    return loss
```

```python
def gradient_descent(x_train,y_train,w,b):
    n= 100000
    for k in range(n):
        for i in range(8):
            dw = 0
            db = 0
            
            for j in range(1000):
                func = np.dot(x_train[j],w)+ b
                dwtemp = (y_train[j] - func) * x_train[j,i]
                dbtemp = (y_train[j] - func)
                dw += dwtemp
                db += dbtemp
            dw[i] /= len(x_train[:,i])
            db[i] /= len(x_train[:,i])
        w = w - 0.001*dw
        b = b - 0.001*db
            
    return w,b
```

```python
x_train = x_train.astype(np.float64)
x_train,x_std,x_mean = z_score(x_train)

np.random.seed(2147483647)
w = np.random.randn(x_train.shape[1],1)
b = np.random.randn(1)

old_cost = 0

while abs(old_cost - cost(x_train,y_train,w,b))>0.00001:
  old_cost = cost(x_train,y_train,w,b)
  w,b = gradient_descent(x_train,y_train,w,b)

x_predict = pd.read_excel('Test data.xlsx').iloc[:,:8].to_numpy()
x_predict = feature_changing(x_predict)
x_predict = (x_predict - x_mean)/x_std
ans = pd.read_excel('Test data.xlsx').iloc[:,8].to_numpy()

y_predict = np.dot(x_predict,w) + b

accuracy = 0
for dim in range(len(ans)):
  if abs(y_predict[dim]-ans[dim])<0.5: # do not change the tolerance as you'll be checked on +- 0.5 error only
    accuracy += 1
accuracy = round(accuracy*100/200.0,2)
ok = 'Congratulations' if accuracy>95 else 'Optimization required'
print(f"{ok}, your accuracy is {accuracy}%")
```
