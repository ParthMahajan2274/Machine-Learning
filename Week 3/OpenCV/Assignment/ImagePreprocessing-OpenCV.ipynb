### Cartoon
For the bart image, you need to mask the dress. Write all your code in the cell below and the final image should be stored in ```cartoon_y_mentee``` variable only or it'll show error later.

````python
#START CODE HERE (Do all your processing here)

cartoon_hsv = cv2.cvtColor(cartoon,cv2.COLOR_BGR2HSV)
lower=np.array([40, 100,100])
upper=np.array([75, 255, 255])

# lower & upper found using https://youtu.be/x4qPhYamRDI?si=fA40RMoFguWzE8iZ

cartoon_rgb = cv2.cvtColor(cartoon,cv2.COLOR_BGR2RGB)

mask = cv2.inRange(cartoon_hsv,lower,upper)
# cv2.imshow('mask',mask)

cartoon_edit = cv2.bitwise_and(cartoon,cartoon,mask=mask)
# bitwise_and got from https://www.tutorialspoint.com/how-to-mask-an-image-in-opencv-python#:~:text=We%20can%20apply%20a%20mask,of%20color%20values%20in%20HSV.

cartoon_y_mentee = cv2.cvtColor(cartoon_edit,cv2.COLOR_BGR2RGB)#(in BGR format only)
# cv2.imshow('cartoon_y_mentee',cartoon_y_mentee)
#END CODE HERE (the final processed image should be cartoon_y_mentee)
````
